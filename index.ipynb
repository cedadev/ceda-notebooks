```python
from notebooks.mngmt.utils.venv_utils import *

venv_name = 'auto_notebook_env'
packages = []
with open('requirements.txt', 'r') as reader:
    packages = [line.strip() for line in reader.readlines()]
    
setup_venv(venv_name, packages)


import ipywidgets as widgets
import simplejson as json
from notebooks.mngmt.utils.index_utils import *

text_content = widgets.Text()
out = widgets.Output(layout={'overflow':'scroll hidden', 'width':'2000px', 'diplay':'flex'})
index_dict = {}

with open('index.json', 'r') as reader:
    index_dict = json.load(reader)

with out:
    print_index(index_dict, display_format='html')


@out.capture(clear_output=True)
def search(event):
    search_index(text_content.value, index_dict, display_format='html')
    
        
button = widgets.Button(description='click here')
button.on_click(search)

widgets.VBox([widgets.HBox([button, text_content]), out])
```
